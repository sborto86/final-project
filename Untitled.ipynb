```python
mean = news["ukraine"].mean()
```

```python
mean
```

```python
dates = []
for i in range (0,len(news)-7):
    start = news.index[i] - datetime.timedelta(days=3)
    end = news.index[i] + datetime.timedelta(days=3)
    test = news[(news.index > start) & (news.index < end)]
    time = pd.DatetimeIndex(news.index).date
    if test["ukraine"].mean() > 3*mean:
        dates.append(time[i])
```

```python
periods = []
for d in dates:
    included = False
    for p in periods:
        if p[0] <= d and p[1] >= d:
            included = True
    if not included:
        start = d - datetime.timedelta(days=2)
        end = d
        while end in dates:
            end = end + datetime.timedelta(days=1)
        periods.append((start,end))
```
